```python
import pandas as pd
```

```python
monthly_expenses_sheet_id = '10OGbaywwMIqKgnPGy8VDvpBVtjyqln47iYa2lFhI9Mw'
monthly_expenses = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{monthly_expenses_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
monthly_expenses
```

```python
last_month_expense = monthly_expenses[monthly_expenses['month'] == monthly_expenses['month'].max()]
last_month_expense
```

```python
crm_expense = last_month_expense[last_month_expense['item'] == 'Salesforce']['amount'].values[0]
crm_expense
```

```python
TotalExpense = crm_expense
```

```python
payroll_sheet_id = '1c_WihqTZCQvNgxzmd-OwhR9i5diwtfxXVLyMn8R-Lp4'
payroll_monthly = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{payroll_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
payroll_monthly
```

```python
last_month_payroll = payroll_monthly[payroll_monthly['month'] == payroll_monthly['month'].max()]
last_month_payroll
```

```python
last_month_payroll_sale_mkt_expense = last_month_payroll[last_month_payroll['department'].isin(['Sales', 'Marketing'])]['paid'].sum()
last_month_payroll_sale_mkt_expense
```

```python
TotalExpense += last_month_payroll_sale_mkt_expense
```

```python
marketing_spendings_sheet_id = '1AZOIThOV4P-0eYDge53ZwumVkfkHoYPWxst3k3Bv87c'
marketing_spendings = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{marketing_spendings_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
marketing_spendings
```

```python
last_month_marketing_expense = marketing_spendings[marketing_spendings['date'].dt.month == marketing_spendings['date'].max().month]['spending'].sum()
last_month_marketing_expense
```

```python
TotalExpense += last_month_marketing_expense
```

```python
TotalExpense
```

```python
receipt_history_sheet_id = '1qayqML1zCKdmtzutkcy9LWvE6xFRm6TGBEVkHHJKIuE'
receipt_history = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{receipt_history_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
receipt_history
```

```python
last_month_new_customer = receipt_history[(receipt_history['date'].dt.month == receipt_history['date'].max().month) & (receipt_history['new_customer'] == 1)]
last_month_new_customer
```

```python
number_customer = len(last_month_new_customer)
number_customer
```

#1. CAC: Customer Acquisition Cost

```python
CAC = TotalExpense/number_customer
CAC
```

#2. ARPU: Average Revenue Per User

```python
receipt_history
```

```python
total_revenue = receipt_history[(receipt_history['date'].dt.month == receipt_history['date'].max().month)]['receipt_amount'].sum()
total_revenue
```

```python
number_customer_last_month = receipt_history[(receipt_history['date'].dt.month == receipt_history['date'].max().month)]['customer_id'].nunique()
number_customer_last_month
```

```python
ARPU = total_revenue/number_customer_last_month
ARPU
```

#3.COGS: Cost of Goods Sold

```python
monthly_expenses
```

```python
production_item = ['AWS Hosting', 'Google Cloud Storage', 'Atlassian Jira']
last_month_production_expense = last_month_expense[last_month_expense['item'].isin(production_item)]['amount'].sum()
last_month_production_expense
```

```python
shared_software_item = ['Zoom', 'Slack']
last_month_share_software_item = last_month_expense[last_month_expense['item'].isin(shared_software_item)]['amount'].sum()
last_month_share_software_item
```

```python
server_n_software_exprense = last_month_production_expense + last_month_share_software_item * 0.6
server_n_software_exprense
```

```python
last_month_production_salary = last_month_payroll[last_month_payroll['department']=='Engineering']['paid'].sum()
last_month_production_salary
```

```python
COGS = server_n_software_exprense + last_month_production_salary
COGS
```

#4.Gross Margin

```python
GrossMargin = (total_revenue - COGS)/total_revenue * 100
GrossMargin
```

#5.LTV: Customer Lifetime Value

```python
life_sheet_id = '1by8tPHwOnq3uKYK2E7sA9VBUYoPM4p1Rnrm_Ss9cyHI'
life_data = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{life_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
life_data
```

```python
life_data['lifespan_day'] = (life_data['churn_date'] - life_data['start_date']).dt.days
life_data
```

```python
avg_lifespan_days = life_data['lifespan_day'].mean()
avg_lifespan_month = avg_lifespan_days / 30
```

```python
LTV = ARPU * avg_lifespan_month * (GrossMargin/100)
LTV
```

#6.LTV / CAC

```python
LTV_CAC = LTV / CAC
LTV_CAC
```
